```python
# Predict Canada's per capita income in year 2020 using canada_per_capita_income.csv
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#load data frame
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
%matplotlib inline
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='.')
plt.show()
```

```python
# splitting x axis
x_years = df.drop('per capita income (US$)',axis='columns')
x_years.head(5)
```

```python
# splitting y axis
y_PCINC_US = df['per capita income (US$)']
y_PCINC_US.head(5)
```

```python
# creating linear regression object
reg = linear_model.LinearRegression()
reg.fit(x_years,y_PCINC_US)
```

```python
# predicting canada's per capita income in year 2020
pred_2020 = reg.predict(pd.DataFrame([[2020]], columns=['year']))
print('per capita income in year 2020: ', pred_2020.flatten().tolist()[0])
```
